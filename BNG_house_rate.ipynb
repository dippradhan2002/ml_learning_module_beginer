# FEATURE SELECTION

DROPPING CONSTANT

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
def drop_constant_features(df, threshold=0.0):
    # Initialize VarianceThreshold with the given threshold
    selector = VarianceThreshold(threshold=threshold)
    # Apply the selector to the DataFrame
    selector.fit(df.select_dtypes(include=['int64', 'float64']))
    # Get the mask of features to keep
    features_to_keep = selector.get_support(indices=True)
    # Get the names of the features to keep
    feature_names = df.select_dtypes(include=['int64', 'float64']).columns[features_to_keep]
    # Return the DataFrame with constant features removed
    return df[feature_names]
```

```python
df = drop_constant_features(df)
```

```python
def correlation (dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) > threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

MUTUAL DEPENDENCY

```python
from sklearn.feature_selection import SelectPercentile, mutual_info_regression
```

```python
# Calculate mutual information
mutual_info = mutual_info_regression(X_train, y_train)
# Convert to a Series and assign column names
mutual_info = pd.Series(mutual_info, index=X_train.columns)
# Sort mutual information values in descending order
mutual_info_sorted = mutual_info.sort_values(ascending=False)
# Plot the mutual information values
mutual_info_sorted.plot.bar(figsize=(15,5))
plt.show()
```

```python
# Initialize SelectPercentile with mutual_info_regression and percentile=30
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=30)
# Fit and transform X_train to select the top columns
X_train_selected = selected_top_columns.fit_transform(X_train, y_train)
# Transform X_test to select the same columns as in X_train
X_test_selected = selected_top_columns.transform(X_test)
# Get the selected column names
selected_columns = X_train.columns[selected_top_columns.get_support()]
# Create new DataFrames with selected columns
X_train_selected_df = pd.DataFrame(X_train_selected, columns=selected_columns)
X_test_selected_df = pd.DataFrame(X_test_selected, columns=selected_columns)
```

```python
print("Selected columns:")
print(selected_columns)
```

```python
# Calculate mutual information
mutual_info = mutual_info_regression(X_train_selected_df, y_train)
# Convert to a Series and assign column names
mutual_info = pd.Series(mutual_info, index=X_train_selected_df.columns)
# Sort mutual information values in descending order
mutual_info_sorted = mutual_info.sort_values(ascending=False)
# Plot the mutual information values
mutual_info_sorted.plot.bar(figsize=(15,5))
plt.show()
```
